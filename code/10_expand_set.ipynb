```python
bi_co_sa = pd.read_csv("{}01_comparison_sample.tsv".format(PAO), sep="\t", index_col=0)

nu_fe_sa = pd.read_csv("{}number_feature_sample.tsv".format(PAO), sep="\t", index_col=0)

se_fe_ = read_set()

en_se_sa = pd.read_csv("{}enrichment_set_sample.tsv".format(PAO), sep="\t", index_col=0)
```

```python
for co, ta in bi_co_sa.iterrows():

    ta.dropna(inplace=True)

    fu = kwat.information.get_ic

    st = read_statistic(PAO, kwat.path.clean(co), fu.__name__)

    pac = "{}expand_set/{}/".format(PAO, kwat.path.clean(co))

    kwat.path.make(pac)

    for se in SETTING["se_"]:

        pas = "{}{}/".format(pac, kwat.path.clean(se))

        kwat.path.make(pas)

        kwat.function_heat_map.make(
            ta,
            pd.concat(
                [
                    nu_fe_sa.loc[nu_fe_sa.index.intersection(se_fe_[se])],
                    en_se_sa.loc[[se]],
                ]
            ),
            st,
            n_pl=None,
            tyt="binary",
            st=SETTING["st"],
            title=se,
            pa=pas,
        )
```
