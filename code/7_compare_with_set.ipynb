```python
bi_co_sa = pd.read_csv("../output/01_comparison_sample.tsv", sep="\t", index_col=0)

en_se_sa = pd.read_csv("../output/enrichment_set_sample.tsv", sep="\t", index_col=0)
```

```python
bu_se_fe_ = {}

pa = "../input/set/"

for bu in os.listdir(path=pa):

    bu_se_fe_[bu] = kwat.gmt.read(["{}{}".format(pa, bu)])
```

```python
fu = kwat.information.get_ic

fun = fu.__name__

ke = {
    "tyt": "binary",
    "st": SETTING["st"],
}

for co, ta in bi_co_sa.iterrows():

    ta.dropna(inplace=True)

    pac = "../output/compare_set/{}/{}/".format(kwat.path.clean(co), fun)

    kwat.path.make(pac)

    st = kwat.function_heat_map.make(
        ta,
        en_se_sa,
        fu,
        n_jo=SETTING["n_jo"],
        n_sa=SETTING["n_sa"],
        n_sh=SETTING["n_sh"],
        n_pl=SETTING["n_ex"],
        title="All ({})".format(fun),
        pa=pac,
        **ke,
    )

    if 0 < len(SETTING["se_"]):

        kwat.function_heat_map.make(
            ta,
            en_se_sa.reindex(labels=SETTING["se_"]),
            st,
            n_pl=None,
            title="Peek ({})".format(fun),
            **ke,
        )

        plot_peek(st["Score"], SETTING["se_"], pac)

    for bu, se_fe_ in bu_se_fe_.items():

        pab = "{}{}/".format(pac, bu)

        kwat.path.make(pab)

        se_ = en_se_sa.index.intersection(se_fe_)

        if 0 < len(se_):

            kwat.function_heat_map.make(
                ta,
                en_se_sa.loc[se_, :],
                st,
                n_pl=SETTING["n_ex"],
                title="{} ({})".format(bu, fun),
                pa=pab,
                **ke,
            )
```
