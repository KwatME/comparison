```python
bi_co_sa = pd.read_csv("{}01_comparison_sample.tsv".format(PAO), sep="\t", index_col=0)

nu_fe_sa = pd.read_csv("{}number_feature_sample.tsv".format(PAO), sep="\t", index_col=0)
```

```python
ke = {
    "tyt": "binary",
    "st": SETTING["st"],
}

for co, ta in bi_co_sa.iterrows():

    ta.dropna(inplace=True)

    for fu in [
        separate_and_get_difference,
        kwat.information.get_ic,
    ]:

        fun = fu.__name__

        pa = "{}compare_feature/{}/{}/".format(PAO, kwat.path.clean(co), fun)

        kwat.path.make(pa)

        st = kwat.function_heat_map.make(
            ta,
            nu_fe_sa,
            fu,
            n_jo=SETTING["n_jo"],
            n_sa=SETTING["n_sa"],
            n_sh=SETTING["n_sh"],
            n_pl=SETTING["n_ex"],
            title="All ({})".format(fun),
            pa=pa,
            **ke,
        )

        if 0 < len(SETTING["fe_"]):

            kwat.function_heat_map.make(
                ta,
                nu_fe_sa.reindex(labels=SETTING["fe_"]),
                st,
                n_pl=None,
                title="Peek ({})".format(fun),
                **ke,
            )

            plot_peek(st.iloc[:, 0], SETTING["fe_"], pa)
```
