```python
nu_fe_sa = pd.read_csv("../output/number_feature_sample.tsv", sep="\t", index_col=0)

bi_co_sa = pd.read_csv("../output/01_comparison_sample.tsv", sep="\t", index_col=0)

se_fe_ = kwat.gmt.read(kwat.path.list("../input/set/"))

sc_se_sa = pd.read_csv("../output/score_set_sample.tsv", sep="\t", index_col=0)
```

```python
for co, ta in bi_co_sa.iterrows():

    ta.dropna(inplace=True)

    fu = kwat.information.get_ic

    st = read_statistic(kwat.path.clean(co), fu.__name__)

    pa = "../output/expand_set/{}/".format(kwat.path.clean(co))

    kwat.path.make(pa)

    for se in SETTING["se_"]:

        pa2 = "{}{}/".format(pa, kwat.path.clean(se))

        kwat.path.make(pa2)

        kwat.function_heat_map.make(
            ta,
            pd.concat(
                [
                    nu_fe_sa.loc[nu_fe_sa.index.intersection(se_fe_[se])],
                    sc_se_sa.loc[[se]],
                ]
            ),
            st,
            n_pl=None,
            tyta="binary",
            st=SETTING["st"],
            title=se,
            pa=pa2,
        )
```
