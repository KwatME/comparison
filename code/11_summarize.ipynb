```python
pd.concat([nu_fe_sa, en_se_sa])
```

```python
for (co, ta), bl_an_ in zip(bi_co_sa.iterrows(), SETTING["su_"]):

    ta.dropna(inplace=True)

    fu = kwat.information.get_ic

    pa = "{}summarize/{}/".format(PAO, kwat.path.clean(co))

    kwat.path.make(pa)

    bu_ = []

    bl_an_["Peek"] = SETTING["fe_"] + SETTING["se_"]

    for bl, an_ in bl_an_.items():

        bu_.append(
            {
                "name": bl,
                "data": pd.concat([nu_fe_sa, en_se_sa]).reindex(labels=an_),
                "type": "continuous",
                "statistic": read_statistic(PAO, kwat.path.clean(co), fu.__name__),
            }
        )

    kwat.function_heat_map.summarize(
        ta,
        bu_,
        ty="binary",
        st=SETTING["st"],
        title=title,
        pa="{}{}.html".format(pa, kwat.path.clean(title)),
    )
```
