```python
bi_co_sa = pd.read_csv("../output/01_comparison_sample.tsv", sep="\t", index_col=0)

nu_fe_sa = pd.read_csv("../output/number_feature_sample.tsv", sep="\t", index_col=0)

en_se_sa = pd.read_csv("../output/enrichment_set_sample.tsv", sep="\t", index_col=0)
```

```python
title = "Summary"
```

```python
for (co, ta), bl_se_ in zip(bi_co_sa.iterrows(), SETTING["su_"]):

    ta.dropna(inplace=True)

    fu = kwat.information.get_ic

    pa = "../output/summarize_set/{}/".format(kwat.path.clean(co))

    kwat.path.make(pa)

    bu_ = []

    for bl, se_ in bl_se_.items():

        bu_.append(
            {
                "name": bl,
                "data": pd.concat([nu_fe_sa, en_se_sa]).reindex(labels=se_),
                "type": "continuous",
                "statistic": read_statistic(kwat.path.clean(co), fu.__name__),
            }
        )

    kwat.function_heat_map.summarize(
        ta,
        bu_,
        ty="binary",
        st=SETTING["st"],
        title=title,
        pa="{}{}.html".format(pa, kwat.path.clean(title)),
    )
```
